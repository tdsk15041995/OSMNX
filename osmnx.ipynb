### Open Street Map(OSM) Network for Mumbai

```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
G = ox.graph_from_place('Mumbai', network_type='drive')
ox.plot_graph(G)
```

### Options available with network type 

drive : drivable public streets (not service roads)

drive_service : drivable streets including service roads

walk : streets and paths that pedestrains can use

bike : streets and paths that cyclist can be

all : all non private OSM streets and paths

all_private : all streets and paths

```python
gulmohar_location = (19.129702,72.9151025)
```

```python
G1 = ox.graph_from_point(gulmohar_location,distance=500,simplify=False,network_type='all')
ox.plot_graph(G1,node_color='orange',node_zorder=2)
```

```python
G1 = ox.graph_from_point(gulmohar_location,distance=500,simplify=False,network_type='drive')
ox.plot_graph(G1,node_color='orange',node_zorder=2)
```

```python
G1 = ox.graph_from_point(gulmohar_location,distance=500,simplify=False,network_type='drive_service')
ox.plot_graph(G1,node_color='orange',node_zorder=2)
```

```python
nodes = [node for node in G1.nodes()]
print("No of Nodes:",len(nodes))
```

```python
edges = [edge for edge in G1.edges()]
print("No of Edges:",len(edges))
```

### What if simplify = True?

```python
G2 = ox.graph_from_point(gulmohar_location,distance=500,simplify=True,network_type='drive_service')
ox.plot_graph(G2,node_color='orange',node_zorder=2)
```

```python
nodes = [node for node in G2.nodes()]
print("No of nodes:",len(nodes))
```

```python
edges = [edge for edge in G2.edges()]
print("No of edges:",len(edges))
```

### Nearest Edge

```python
G3 = ox.graph_from_point(gulmohar_location,distance=500,simplify=False,network_type='drive_service')
fig,ax=ox.plot_graph(G3,node_color='orange',node_zorder=2,show=False,close=False)
ax.scatter(gulmohar_location[1],gulmohar_location[0],c='red',s=20)
plt.show()
```

```python
tup=ox.geo_utils.get_nearest_edge(G3,gulmohar_location)
```

```python
nc = ['green' if node == tup[1] or node == tup[2] else 'orange' for node in G3.nodes()]
```

```python
fig,ax=ox.plot_graph(G3,node_color=nc,node_zorder=2,show=False,close=False)
ax.scatter(gulmohar_location[1],gulmohar_location[0],c='red',s=20)
plt.show()
```
