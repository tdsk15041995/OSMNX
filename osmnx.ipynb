```python
import osmnx as ox
G = ox.graph_from_place('Mumbai', network_type='drive')
ox.plot_graph(G)
```

```python
location = (19.1215092,72.9105231)
G1 = ox.graph_from_point(location,distance=500,simplify=False)
ox.plot_graph(G1,node_color='orange',node_zorder=3)
```

```python
nodes = [node for node in G1.nodes()]
print(len(nodes))
```

```python
edges = [edge for edge in G1.edges()]
print(len(edges))
```

```python
nc = ['b' if ox.is_endpoint(G1,node) else 'r' for node in G1.nodes()]
ox.plot_graph(G1,node_color=nc)
```

```python
G2 = G1.copy()
G2 = ox.simplify_graph(G1)
ox.plot_graph(G2)
```

```python
nodes = [node for node in G2.nodes()]
len(nodes)
```

```python
edges = [edge for edge in G2.edges()]
len(edges)
```
